# Clustering customer income and spendings

```python
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('Mall_Customers.csv')
df.head(3)
```

Kmeans model wordt getraind an predict de clusters 

```python
df = df.iloc[:, [3,4]]
kmeansmodel = KMeans(n_clusters= 5, init='k-means++', random_state=0)
y_kmeans = kmeansmodel.fit_predict(df.values)
```

Kijk welke labels er allemaal mogelijk zijn om toegepast te worden voor de scatterplot (0, 1, 2, 3, 4)

```python
y_kmeans
```

Plot de scatterplot en identificeer de clusters die gevonden zijn

```python
df = df.values

# tweede "0" in "df[y_kmeans == 1, 0]" staat voor de kolom de waarde uit moet komen.
plt.scatter(df[y_kmeans == 0, 0], df[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1') # first cluster with 0
plt.scatter(df[y_kmeans == 1, 0], df[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2') # second cluster with only 1s
plt.scatter(df[y_kmeans == 2, 0], df[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3') # thirds cluster with only 2s
plt.scatter(df[y_kmeans == 3, 0], df[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') # fourth cluster with only 3s
plt.scatter(df[y_kmeans == 4, 0], df[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') # fifth cluster with only 4s

# plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
