# Z-scores Airbnb prices New York

```python
from scipy import stats

import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('AB_NYC_2019.csv')
df.head(3)
```

```python
# Calculates the Zscore of a value
def z_score(value, mean, std):
    return (value - mean) / std

mean = np.mean(df.price)
std = np.std(df.price)
```

```python
print('mean: ', mean)
print('std:', std)

print('value - z_score')
print(df.price[0],'_', z_score(df.price[0], mean, std))
print(df.price[1],'_', z_score(df.price[1], mean, std))
print(df.price[2],'_', z_score(df.price[2], mean, std))
print(df.price[3],'_', z_score(df.price[3], mean, std))
print(df.price[4],'_', z_score(df.price[4], mean, std))
```

```python
z_scores = stats.zscore(df.price)
import matplotlib.pyplot as plt
plt.hist(df.price, bins = 100, range=(0, 1000))
sns.kdeplot(z_scores)
```

# Evaluatie
Aangezien de standaard deviatie ver boven het gemiddelde van de prijzen ligt, kan erg geconstateerd worden dat de prijzen een groot verschil onderling hebben. Hoe lager de standaard deviatie is hoe minder variatie er in de waardes van de dataset zitten, dit duidt waarschijnlijk ook vaak op een dataset die geen goede normaalverdeling heeft. Dit komt doordat veel waardes zitten die boven het gemiddelde zitten, waardoor er dus een grafiek zoals in de histogram hierboven onstaat met waardes die oplopen ver boven het gemiddelde.
